```python
df = pd.read_excel('localizacao_vargem.xlsx')
df
```

```python
def cadastra_localizacao(df):
    driver = webdriver.Chrome()#options=options)
    driver.get(f"http://{link_ip}:4647/sgfpod1/Cad_0020.pod")
    sleep(1)
    driver.find_element(By.NAME, 'id_cod_usuario').send_keys(login_trier)
    sleep(1)
    driver.find_element(By.NAME, 'nom_senha').send_keys(password_trier)
    sleep(1)
    driver.find_element(By.NAME, 'login').click()
    sleep(1)
    erro = []
    for index, row in tqdm(df.iterrows(), total=df.shape[0]):
        try:
            sleep(1)
            driver.refresh()
            sleep(1)
            driver.find_element(By.NAME, 'cod_redbarraEntrada').send_keys(f"{row['ean']}")
            driver.find_element(By.NAME, 'cod_redbarraEntrada').send_keys(Keys.RETURN)
            sleep(1)
            configuration_element = driver.find_element(By.XPATH, "//a[contains(text(), 'Configurações')]")
            configuration_element.click()
            sleep(2)
            # Encontre o elemento pelo atributo 'name' e insira o texto
            loc_element = driver.find_element(By.NAME, "nom_local")
            loc_element.clear()
            loc_element.send_keys(f"{row['localizacao']}")
            sleep(1)
            salvar = driver.find_element(By.NAME, "salvar")
            salvar.click()
            sleep(3)
        except:
            erro.append(row['ean'])
            continue
    return erro
```

```python
erro = cadastra_localizacao(df)
erro
```
