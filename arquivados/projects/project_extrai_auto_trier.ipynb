```python
from time import sleep
from datetime import date
from selenium import webdriver
from selenium.webdriver.chrome.service import Service 
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_excel('clientes_sintraturb.xls')
df_sintraturb = pd.DataFrame()
df_sintraturb['cod'] = df['Cód.'][:-1]
df_sintraturb['cliente'] = df['Cliente'][:-1]
df_sintraturb
```

```python
df_sintraturb = df_sintraturb[180:]
df_sintraturb
```

```python
texto_obs = '''### ATENCAO ### SOMENTE E PERMITIDO A VENDA DE MEDICAMENTOS, FRALDAS OU LEITES PARA ESSE CLIENTE NO CONVENIO DO SINTRATURB. PARA VENDAS DE NAO-MEDICAMENTOS, REALIZAR OUTRA VENDA EM DINHEIRO OU CARTAO!'''
```

```python
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
driver.implicitly_wait(2)
url = 'http://179.184.16.200:4647/sgfpod1/Menu.pod'
driver.get(url)
sleep(1)
driver.find_element(by=By.NAME,value='id_cod_usuario').send_keys(119)
driver.find_element(by=By.NAME,value='nom_senha').send_keys(190224)
sleep(1)
driver.find_element(by=By.NAME,value='login').click()
sleep(2)

url = 'http://179.184.16.200:4647/sgfpod1/Cad_0002.pod'
driver.get(url)
sleep(5)

for cod_cliente in tqdm(df_sintraturb['cod'], desc="Atualizando clientes", unit="cliente"):
    driver.find_element(by=By.ID,value='id_cod_clienteEntrada').send_keys(cod_cliente)
    driver.find_element(by=By.ID,value='id_cod_clienteEntrada').send_keys(Keys.RETURN)
    sleep(1)
    driver.find_element(by=By.LINK_TEXT,value='Inf. Adicionais').click()
    sleep(1)
    driver.find_element(by=By.ID,value='nom_observacao').clear()
    driver.find_element(by=By.ID,value='nom_observacao').send_keys(texto_obs)
    driver.find_element(by=By.ID,value='btSalvarGeral').click()
    sleep(2)

```
